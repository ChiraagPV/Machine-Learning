# RECOMMENDING RESTAURANTS USING CONTENT BASED RECOMMENDATION SYSTEM

### IN THIS PROJECT WE WILL RECOMMEND RESTAURANTS TO THE USERS BASED ON THEIR INPUTS AND THEIR CHOICE OF CUISINE

```python
import pandas as pd
import numpy as np
```

```python
Restaurant = pd.read_excel('Restaurants.xlsx')
Restaurant = Restaurant.drop('Links',axis = 1)
Restaurant.head()
```

```python
user_inp = pd.read_excel('User_inp.xlsx')
user_inp = user_inp.drop('Links',axis = 1)
user_inp
```

```python
Restaurant.describe(include='all').T
```

```python
user_inp.describe(include='all').T
```

```python
Restaurant['Ratings'] = Restaurant['Ratings'].replace('New','0').replace('-','0').astype('float')
Restaurant.Ratings.unique()
```

```python
user_inp['Ratings'] = user_inp['Ratings'].replace('New','0').replace('-','0').astype('float')
user_inp.Ratings.unique()
```

```python
X = Restaurant[['Restaurant Names','Ratings','Cuisine']]
XX = X
X.head()
```

```python
Y = user_inp[['Restaurant Names','Ratings','Cuisine']]
YY = Y
Y.head()
```

```python
X['Cuisine'] = X.Cuisine.str.split(',')
Y['Cuisine'] = Y.Cuisine.str.split(',')
```

```python
XX = X.copy()

try:
    for index, row in X.iterrows():
        for cuisine in row['Cuisine']:
            XX.at[index, cuisine] = 1
except:
    print('')

XX = XX.fillna(0)
XX.head()
```

```python
YY = Y.copy()

try:
    for index, row in Y.iterrows():
        for cuisine in row['Cuisine']:
            YY.at[index, cuisine] = 1
except:
    print('')

YY = YY.fillna(0)
YY.head()
```

## We will extract cuisines and ratings of the restaurants into 2 different dataframes

We are selecting specific cuisines given by users as inputs.

```python
cuis = XX.drop('Restaurant Names',axis = 1).drop('Ratings',axis = 1).drop('Cuisine',axis = 1)
cuisine = cuis[['Chinese', ' Bakery', ' Sichuan', ' Pizza', ' Burger', 'Burger',
       ' Fast Food', ' Biryani', ' Desserts', ' Beverages', 'South Indian',
       ' Andhra', ' Mithai', 'Biryani', ' Kebab', 'Fast Food', ' Sandwich',
       ' Wraps', ' Rolls', ' Salad', 'North Indian', ' Chinese', ' Mughlai',
       ' Mandi', ' Shawarma', ' Seafood', ' North Indian', 'Ice Cream']]
print(cuisine.shape)
cuisine.head()
```

```python
ratings = XX[['Ratings']]
ratings.head()
```

```python
user_cuisine = YY.drop('Restaurant Names',axis = 1).drop('Ratings',axis = 1).drop('Cuisine',axis = 1)
print('USER INPUTS : Cuisines rows and columns : ',user_cuisine.shape)
user_cuisine.head()
```

```python
user_cuisine.columns
```

```python
user_ratings = YY[['Ratings']]
user_ratings.head()
```

## We have collected all the possible data : 
### cuisine, ratings, user_cuisine, user_ratings

# CONTENT BASED RECOMMENDATION SYSTEM

STEP 1 : LETS BUILD A USER PROFILE

a - user_ratings * user_cuisine <br>
b - Compute weighted matrix by finding the sum of each columns (ex : sum of entire North Indian column)

A USER PROFILE IS BUILT : user_profile

```python
user_r = user_ratings.to_numpy()
user_c = user_cuisine.to_numpy()
print(user_r.shape, user_c.shape)
```

```python
rc = user_r * user_c
rc[:1]
```

```python
weighted_matrix = rc.sum(axis=0)
print(weighted_matrix)
print(weighted_matrix.shape)
user_profile = weighted_matrix
```

STEP 2 : WE HAVE TO MULTIPLY USER PROFILE WITH THE CUISINE DATA.

a - user_profile * cuisines<br>
b - Compute weighted matrix by finding the sum of each rows (ex : sum of entire (Pizza Hut) restauant row )

```python
cuisines = cuisine.to_numpy()
cuisines[:1]
```

```python
recommend = user_profile * cuisines
recommend[:1]
```

```python
weighted_avg_recommend = recommend.sum(axis = 1)
weighted_avg_recommend.shape
```

```python
final = pd.DataFrame(weighted_avg_recommend)
final = final.rename(columns = {0:'Recommended ratings of restaurants for Users based on User profile'})
X['Recommended ratings of restaurants for Users based on User profile'] = final
```

```python
sorted_X = X.sort_values(by='Recommended ratings of restaurants for Users based on User profile',ascending = False)
```
